```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from matplotlib import pyplot as plt

from dtreeviz import model as dtreeviz_model

import graphviz.backend as be

from IPython.display import Image, display_svg, SVG

import warnings
import logging

import numpy as np
```

```python
iris = load_iris()

x = iris.data
y = iris.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
dt = DecisionTreeClassifier(criterion='gini', max_depth = 4)
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy on test set is {accuracy:.2f}")
```

```python
plot_tree(dt, filled = True, feature_names = iris.feature_names, class_names = iris.target_names)

plt.title("Decision Tree for Iris Dataset")
plt.show()
```

```python
viz = dtreeviz_model(dt,
               x_train,
               y_train,
               target_name = 'species',
               feature_names = iris.feature_names,
               class_names = iris.target_names)

warnings.filterwarnings("ignore")
logging.getLogger('matplotlib').setLevel(logging.ERROR)

viz.view()
```

```python
x_sample = x_test[np.random.randint(0, x_test.shape[0])]

viz = dtreeviz_model(dt,
               x_train,
               y_train,
               target_name = 'species',
               feature_names = iris.feature_names,
               class_names = iris.target_names)

warnings.filterwarnings("ignore")
logging.getLogger('matplotlib').setLevel(logging.ERROR)

viz.explain_prediction_path(x_sample)
```
